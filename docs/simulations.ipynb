```python
import numpy as np
from numpy import random
import pandas as pd
```

# Visual search model

```python

mus = [0,1]; # distractor,target
sds_dict = {'O':[0.1,1],'Q':[1,0.1]}; # distractor,target

responses = [];
search_time = [];
target_present = [];
set_size = [];
correct = [];
target = [];

def create_search_array(target_present,set_size):
    return [0]*(set_size)+[target_present]

def visual_search_trial(items, target):
    
    evidence = [0]
    
    sds = sds_dict[target];
    
    # set threshold to control error rate
    expected_distractor_var = sds[0]**2;
    expected_target_var = sds[1]**2;
    expected_var = expected_distractor_var*len(items)
    expected_sd = np.sqrt(expected_var)
    
    threshold_no = [-expected_sd];
    threshold_yes = [expected_sd];

    while True:

        step = np.array([random.normal(mus[item],sds[item]) for item in items]).sum()

        evidence.append(evidence[-1]+step)

        expected_signal = mus[0]*len(items)+mus[1]/2-mus[0]/2;

        threshold_no.append(threshold_no[-1]+expected_signal);

        threshold_yes.append(threshold_yes[-1]+expected_signal)

        if evidence[-1]>=threshold_yes[-1]:
            response=1
            break
        elif evidence[-1]<=threshold_no[-1]:
            response=0
            break
    return(response, len(evidence))


for t in ['Q','O']:
    for ss in [1,2,4,8]:
        for tp in [0,1]:
            for repeition in range(2000):

                items = create_search_array(tp,ss);
                response,RT = visual_search_trial(items,t);
                print(response)
                responses.append(response);
                search_time.append(RT);
                target_present.append(tp);
                set_size.append(ss);
                correct.append(1*(tp==response))
                target.append(t)
            
symmetrical_search = pd.DataFrame()            
            

        
```

```python
vs_simulation = pd.DataFrame({'target':target,
                                   'target_present':target_present,
                                  'set_size':set_size,
                                  'response':responses,
                                  'search_time':search_time,
                                  'correct':correct})            
vs_simulation.to_csv('vs_simulation.csv')
```

# Discrimination model


```python
mus = [-1,1]; # FN,FP
sds = [1,0.8] #FN,FP

responses = [];
RT = [];
stimulus = [];
correct = [];

def discrimination_trial(stim, start, threshold):
    
    evidence = [start]
    
    while True:
        
        if evidence[-1]>=threshold:
            response=1
            break
        elif evidence[-1]<=-threshold:
            response=0
            break

        step = random.normal(mus[stim],sds[stim])

        evidence.append(evidence[-1]+step)

    return(response, len(evidence))

def set_params(mus, sds,step_size,target_error,eps):
    
    i=0;
    start_vec = [0];
    threshold_vec = [0.01];
    error1 = [0.7];
    error0 = [0.5];
    
    while abs(error1[-1]-error0[-1])>eps or abs(error1[-1]-target_error)>eps or abs(error1[-1]-target_error)>eps:
        
        i+=1
        if i%2==0:
            print('start: %f'%start_vec[-1])
            if error1[-1]>error0[-1]:
                start_vec.append(start_vec[-1]+step_size)
            else:
                start_vec.append(start_vec[-1]-step_size)
        else:
            print('threshold: %f'%threshold_vec[-1])
            if (error1[-1]+error0[-1])/2>target_error:
                threshold_vec.append(threshold_vec[-1]+step_size/2)
            else:
                threshold_vec.append(threshold_vec[-1]-step_size/2)
                
        correct1 = [];
        correct0 = [];
        
        for repeition in range(1000):
            response1,rt = discrimination_trial(1,start_vec[-1],threshold_vec[-1]);
            correct1.append(1*(1==response1));
            response0,rt = discrimination_trial(0,start_vec[-1],threshold_vec[-1]);
            correct0.append(1*(0==response0));
        
        error1.append(1-np.mean(correct1));
        error0.append(1-np.mean(correct0));
        
        print('error1: %f'% error1[-1])
        print('error2: %f'% error0[-1])

    
    return(threshold_vec, start_vec)

```

```python
threshold, start = set_params(mus,sds,step_size=0.001,target_error=0.05,eps=0.005)
```

```python
disc_simulation = pd.DataFrame({'stimulus':stimulus,
                                  'response':responses,
                                  'RT':RT,
                                  'correct':correct})            
disc_simulation.to_csv('disc_simulation.csv')
```
