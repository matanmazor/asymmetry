# Visual search model

```python

mus = [0,1]; # distractor,target
sds_dict = {'O':[0.1,1],'Q':[1,0.1]}; # distractor,target

responses = [];
search_time = [];
target_present = [];
set_size = [];
correct = [];
target = [];

def create_search_array(target_present,set_size):
    return [0]*(set_size)+[target_present]

def visual_search_trial(items, target):
    
    evidence = [0]
    
    sds = sds_dict[target];
    
    # set threshold to control error rate
    expected_distractor_var = sds[0]**2;
    expected_target_var = sds[1]**2;
    expected_var = expected_distractor_var*len(items)
    expected_sd = np.sqrt(expected_var)
    
    threshold_no = [-expected_sd];
    threshold_yes = [expected_sd];

    while True:

        step = np.array([random.normal(mus[item],sds[item]) for item in items]).sum()

        evidence.append(evidence[-1]+step)

        expected_signal = mus[0]*len(items)+mus[1]/2-mus[0]/2;

        threshold_no.append(threshold_no[-1]+expected_signal);

        threshold_yes.append(threshold_yes[-1]+expected_signal)

        if evidence[-1]>=threshold_yes[-1]:
            response=1
            break
        elif evidence[-1]<=threshold_no[-1]:
            response=0
            break
    return(response, len(evidence))


for t in ['Q','O']:
    for ss in [1,2,4,8]:
        for tp in [0,1]:
            for repeition in range(2000):

                items = create_search_array(tp,ss);
                response,RT = visual_search_trial(items,t);
                print(response)
                responses.append(response);
                search_time.append(RT);
                target_present.append(tp);
                set_size.append(ss);
                correct.append(1*(tp==response))
                target.append(t)
            
symmetrical_search = pd.DataFrame()            
            

        
```

# Discrimination model


```python
mus = [-1,1]; # FN,FP
sds = [0.8,1] #FN,FP

responses = [];
RT = [];
stimulus = [];
correct = [];

def discrimination_trial(mus, sds, stim, start, threshold):
    
    evidence = [start]
    
    while True:
        
        if evidence[-1]>=threshold:
            response=1
            break
        elif evidence[-1]<=-threshold:
            response=0
            break

        step = random.normal(mus[stim],sds[stim])

        evidence.append(evidence[-1]+step)

    return(response, len(evidence))

def loss(RT,acc):
    return RT/(acc)

def test_acc_rt(mus, sds, start,threshold,loss,N):
    correct = [];
    RT= [];
    stimulus = [];
    for stim in [0,1]:
            for repeition in range(N):
                response,rt = discrimination_trial(mus, sds, stim,start,threshold);
                stimulus.append(stim);
                correct.append(1*(stim==response));
                RT.append(rt);

    return({'start':start,
            'threshold':threshold,
            'error':1-np.mean(correct),
            'RT':np.mean(RT),
            'loss':loss(np.mean(RT),np.mean(correct)),
            'RTdiff':np.mean(np.array(RT)[np.equal(stimulus,1)])-np.mean(np.array(RT)[np.equal(stimulus,0)]),
            'errordiff':np.mean(np.array(correct)[np.equal(stimulus,0)])-np.mean(np.array(correct)[np.equal(stimulus,1)])})
```

```python
def set_params(mus, sds,step_size,loss_function):
    
    i=0;
    start_vec = [0];
    threshold_vec = [2];
    error = [0.7];
    RT = [0.5];
    loss_vec = [];
    
    num_iter = 5000;
    
    for i in range(num_iter):
        
        N= round(1000*(i/num_iter))
        options = [test_acc_rt(mus, sds, start_vec[-1],threshold_vec[-1],loss,N),
                   test_acc_rt(mus, sds, start_vec[-1]-step_size,threshold_vec[-1],loss,N),
        test_acc_rt(mus, sds, start_vec[-1]+step_size,threshold_vec[-1],loss,N),
        test_acc_rt(mus, sds, start_vec[-1],threshold_vec[-1]-step_size,loss,N),
        test_acc_rt(mus, sds, start_vec[-1],threshold_vec[-1]+step_size,loss,N)]
        random.shuffle(options)
        
        
        best_option = options[np.argmin([option['loss'] for option in options])];
        
#         pdb.set_trace()
                
        start_vec.append(best_option['start']);
        threshold_vec.append(best_option['threshold']);
        error.append(best_option['error']);
        RT.append(best_option['RT']);
        loss_vec.append(best_option['loss']);
        
    return(threshold_vec, start_vec, error, RT, loss_vec)

```

```python
mus = [-1,1]; # FN,FP
sds = [0.5,1] #FN,FP

def step_loss(RT,acc):
    if acc<0.95:
        return 1000
    else: 
        return RT

candidate_threshold = np.arange(1,20,1)/10;
candidate_start = np.arange(-19,20,1)/10;

loss_array = np.empty([len(candidate_threshold), len(candidate_start)]);
acc_array = np.empty([len(candidate_threshold), len(candidate_start)]);
RT_array = np.empty([len(candidate_threshold), len(candidate_start)]);
RTdiff_array = np.empty([len(candidate_threshold), len(candidate_start)]);
errordiff_array = np.empty([len(candidate_threshold), len(candidate_start)]);

for i,thres in enumerate(candidate_threshold):
    for j,start in enumerate(candidate_start):
        
        performance = test_acc_rt(mus, sds, start, thres, step_loss, 5000)
        loss_array[i][j] = performance['loss']
        acc_array[i][j] = 1-performance['error']
        RT_array[i][j] = performance['RT']
        RTdiff_array[i][j] = performance['RTdiff']
        errordiff_array[i][j] = performance['errordiff']



```
